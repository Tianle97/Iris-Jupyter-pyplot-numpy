## 6.Fit a straight line

Fit a straight line to the variables petal length and petal width for the whole data set. Plot the data points in a scatter plot with the best fit line shown.

```python
# Find best fit line
# %f = float 浮点数字(用小数点符号)
# % = a % b returns the remainder after dividing a by b
m,c = np.polyfit(petal_length, petal_width, 1)
print("Best fit is m = %f and c = %f" % (m, c))
```

```python
# y = mx + c.
pl.plot(petal_length, petal_width, 'k.', label='iris petal data')
#pl.plot(m, 'b-', label='Slope: %f' % c)
pl.plot(petal_length, m * petal_length + c, 'b-', label='Best fit (Gradient Descent): $%0.1f x + %0.1f$' % (m,c))
pl.xlabel('petal length')
pl.ylabel('petal width')
pl.legend()
pl.show()
```

## 7.Calculate the R-squared value

Calculate the R-squared value for your line above.

```python
# ref: https://github.com/emerging-technologies/emerging-technologies.github.io/blob/master/notebooks/simple-linear-regression.ipynb
# avg = np.mean()
# %6.15f: .15f = 15 digits after the decimal point（小数点后15位数）
rsq = 1.0 - (np.sum((petal_width - m * petal_length - c)**2)/np.sum((petal_width - np.mean(petal_width))**2))
print("The R-squared value is %6.4f" % rsq)

# The same value using numpy.
np.corrcoef(petal_length, petal_width)[0][1]**2
```

## 10.Use gradient descent

Use gradient descent to approximate the best fit line for the petal length and petal width setosa values. Compare the outputs to your calculations above.

Here we create two functions, each taking four parameters.
The first two parameters are arrays with our $x_i$ and $y_i$ data set values.
The second two are our current guesses for $m$ and $c$.

```python
def grad_m(x,y,m,c):
    return -2.0 * np.sum(x * (y - m * x - c))
```

```python
def grad_c(x,y,m,c):
    return -2.0 * np.sum(y - m * x - c)
```

```python
# eta = η. it often is used for the learning rate. This is the distance with which you will 
# reach to get to the next "more optimal" point when taking the gradient.
eta = 0.0001
gdm, gdc = 1.0, 1.0
change = True

while change:
    mnew = gdm - eta * grad_m(petal_length[0:50:1], petal_width[0:50:1], gdm, gdc)
    cnew = gdc - eta * grad_c(petal_length[0:50:1], petal_width[0:50:1], gdm, gdc)
    if gdm == mnew and gdc == cnew:
        change = False
    else:
        gdm, gdc = mnew, cnew
print("m: %20.16f  c: %20.16f" % (gdm, gdc))
```
